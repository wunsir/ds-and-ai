```python
import earthaccess
earthaccess.login()
```

```python
# 搜索覆盖中国区域的2020年VNP46A2数据文件
search_results = earthaccess.search_data(
    short_name="VNP46A2",          # 产品短名称VNP46A2
    cloud_hosted=True,            # 优先搜索云端存储的数据
    bounding_box=(73, 18, 135, 54), # 中国的大致地理边界框 (西, 南, 东, 北)
    temporal=("2020-01-01", "2020-12-31") # 2020年全年的时间范围
)
search_results
```

```python
print(f"找到了 {len(search_results)} 个匹配的文件。")
```
